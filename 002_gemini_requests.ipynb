````python
# Install dependencies
%pip install google-genai python-dotenv

# Load env variables
from dotenv import load_dotenv
load_dotenv()

# Create an API client
from google import genai

client = genai.Client()
model = "gemini-2.5-flash"
# model = "gemini-2.5-pro"

from google.genai import types

def add_user_message(messages, text):
    messages.append({"role": "user", "parts": [{"text": text}]})

def add_assistant_message(messages, text):
    messages.append({"role": "model", "parts": [{"text": text}]})

def chat(messages, system=None, stop_sequences=None):
    params = {
        "model": model,
        "contents": messages,
    }

    config_kwargs = {}
    if system:
        config_kwargs["system_instruction"] = system
    if stop_sequences:
        config_kwargs["stop_sequences"] = stop_sequences

    if config_kwargs:
        params["config"] = types.GenerateContentConfig(**config_kwargs)

    response = client.models.generate_content(**params)
    return response.text
````

````python
messages = []
add_user_message(messages, "Escríbeme en un párrafo de no menos de 100 palabras cuales son los avances mas relevantes sobre las baterías de estado solido")
params = {
    "model": model,
    "contents": messages,
    "config": types.GenerateContentConfig(max_output_tokens=1000)
}

stream = client.models.generate_content_stream(**params)

for event in stream:
    print(event)
````

````python
# Make an initial list of messages
messages = []
system = """
    You are a patient math tutor.
    Do not directly answer a student's questions.
    Guide them to a solution step by step.
    """
# Use a 'while True' loop to run the chatboot forever

while True:
    # Get user input
    user_input = input("> ")
    print(">", user_input)

# Add user input to the list of messages
    add_user_message(messages, user_input)

# Call Gemini with the 'chat' function
    response = chat(messages, system=system)

# Add generated text to the list of messages
    add_assistant_message(messages, response)

# Print the generated text to the console
    print(">", response)
````
